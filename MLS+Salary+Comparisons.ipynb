```python
def fullname(x, y):
    if str(x) == "nan":
        return str(y)
    else:
        return str(x) + " " + str(y)
```

```python
mls_salaries_2017['Name'] = np.vectorize(fullname)(mls_salaries_2017['first_name'], mls_salaries_2017['last_name'])
```

```python
def position_converter(x):
    if x == 'M-F':
        return str('M/F')
    if x == 'F-M':
        return str('F/M')
    else:
        return str(x)
```
